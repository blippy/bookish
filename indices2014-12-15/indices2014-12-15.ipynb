# The Footsie versus the midcaps and AIM

**Abstract** *In this notebook, I take a look at the relationships between the returns on the Footsie, and those for the midcaps and the AIM over the last decade. I found that there is a strong correlation between the Footsie and the midcaps, but a weaker correlation to the AIM. The midcaps have tended to amplify the gains of the Footsie, but the AIM market has lost money over the last decade.*

## Generating the data

In this section I'll show you how I generated the data. You can skip this section if you just want to use the data that I have provided.

```python
# Define the indices an helper rountines
yahoo_codes = ["^FTSE", # Footsie
               "^FTMC", # Mid 250
               "^FTAI" # AIM Index
               ]

def map_codes(func):
    """map a function over the yahoo codes"""
    global yahoo_codes
    return [func(c) for c in yahoo_codes]
```

```python
#Download data for the indices
if False: map_codes(quant.get_decade)
```

```python
# Load, preprocess and save the downloaded data

# Not very elegant. I'm sure there must be a better way
if False:
    # load the data into a list of dataframes
    dfs = { c: quant.load_to_pandas(c) for c in yahoo_codes }
    
    # Determine the minimum and maximum date range across the indices
    min_dates = map_codes(lambda c: dfs[c]['Date'].min())
    min_date = max(min_dates)
    print("Min dates:", min_date, min_dates)
    max_date = min(max_dates)
    max_dates = map_codes(lambda c: dfs[c]['Date'].max())    
    print("Max dates:", max_date, max_dates)
    
    # Subset the data for the min and max dates
    #min_date = '2009' # tweak some startup nonsens
    for c in yahoo_codes:
        df = dfs[c]
        df = df[(df['Date']>= min_date) & (df['Date']<=max_date)]
        dfs[c] = df

    # check the consistency between the dates
    def darray(c): return np.array(dfs[c]['Date'], dtype = "S10")
    da1, da2, da3 = map_codes(darray)
    ds1 = np.setxor1d(da2, da3)
    print("Differences", ds1)
    ds2 = np.setxor1d(da1, da3)
    print("Differences", ds2)
    ds3 = np.setxor1d(da1, da2)
    print("Differences", ds3)
    
    df = pd.DataFrame()
    df['Date'] = dfs["^FTSE"]['Date']
    df['FTSE'] = dfs["^FTSE"]['Close']
    df = pd.merge(dfs["^FTSE"] , dfs["^FTMC"], 
                  left_on='Date', right_on='Date', how='outer', 
                  suffixes = ('_FTSE', '_FTMC'), sort = True)
    df = pd.merge(df , dfs["^FTAI"], 
                  left_on='Date', right_on='Date', how='outer', 
                  suffixes = ('_1', '_FTAI'), sort = True)
    
    # extract the columns, and save
    h5 = h5py.File("indices2014-12-15.h5", "w")
    def h5save(varname, arr): 
        h5.create_dataset(varname, data = arr, compression="gzip")
    dates = np.array(df['Date'], dtype = "S10")
    h5save('date', dates)
    ftses = np.array(df['Close_FTSE'], dtype = "f64")
    h5save('ft100', ftses)
    ftmcs = np.array(df['Close_FTMC'], dtype = "f64")
    h5save('ft250', ftmcs)
    ftais = np.array(df['Close'], dtype = "f64")
    h5save('ftaim', ftais)
    h5.close()
    print(df[['Date', 'Close_FTSE', "Close_FTMC", "Close"]])
```

## Investigating the data

The indices, normalised to 1 at the beginning of the data, show the following performance with time:

```python
import mython.times
sdates = [ mython.times.str_to_date(d.decode()) for d in dates]
plt.plot(sdates, ft250/ft250[0], 'b') # FT250 shown blue
plt.plot(sdates, ft100/ft100[0], 'g') # FT100 shown green
plt.plot(sdates, ftaim/ftaim[0], 'r') # FTAIM shown red
```

### Summary statistics

In the previous section we estimated that each month contained 21 data points. Let us calculate monthly returns for the three indices under this assumption:

```python
def gain(arr): return arr[21::21]/arr[0:-21:21]

g100 = gain(ft100) # monthly gains of the Footsie
g250 = gain(ft250) # monthly gains of the mid-caps
gaim = gain(ftaim) # monthly gains of the AIMs
```

Some basic stats:

```python
m100, m250, maim = np.mean(g100), np.mean(g250), np.mean(gaim)
print("Monthly returns:", m100, m250, maim)
print("Expressed yearly:", m100**12, m250**12, maim**12)
```

Two things leap out at me:
* The FTSE 250 outperformed the FTSE 100 by over 5% pa over nearly a decade
* It you had invested in AIM, you would be losing nearly 1% pa on average

The correlation coefficients are:

```python
print("Coeff between g100 and g250:", np.corrcoef(g100, g250)[0, 1])
print("Coeff between g100 and gaim:", np.corrcoef(g100, gaim)[0, 1])
```

You can see that most of the gains in the FT250 are accounted for by gains in the FT100. In the case of AIM, the relationship is much weaker.

### FT100 and the FT250

Here is a plot of the FT250 against the FT100:

```python
plt.scatter(ft100, ft250)
```

A scatter plot of the gains for the FT250 (represented by g250) against those for the FT100 (represented by g100) is as follows:

```python
def splot(arr1, arr2):
    fig, ax = plt.subplots(nrows = 1, ncols = 1)
    fig.set_size_inches(8, 8)
    ax.scatter(arr1, arr2)
    plt.show()
    
splot(g100, g250)
```

The above plot shows a tight banding between gains in the FT100 and gains in the FT250.

The relationship looks approximately linear, so we can fit a first-order polynomial between them. A fit between the monthly gains for the ft100 (i.e. g100) and those for the ft250 (i.e. g250), taking g100 as the independent variable can be calculated as follows:

```python
np.polyfit(g100, g250, 1)
```

``polyfit`` returns the polynomial coefficients with highest power first ([documentation](http://docs.scipy.org/doc/numpy/reference/generated/numpy.polyfit.html)). So, the monthly returns are fitted as follows:

$g_{250} = -0.01052049 + 1.01505566 \times g_{100}$

This equation shows a negative intercept (I'm not sure how this should be interpreted - it seems counterintuitive to me), but the gradient is greater than 1. There is therefore a multiplier effect. The FT250 looks like it is a leveraged bet on the FT100. Intuitively, this makes sense: large-caps are considered "safe and dependable", whilst the mid-caps are seen as risky.

### FT100 and the AIM

A plot of the AIM index against the FT100 shows a much more chaotic pattern than it did for the FT250:

```python
plt.scatter(ft100, ftaim)
```

A plot of the gains on the AIM against the FT100 shows a positive correlation, but with much greater variability compared to the FT250:

```python
splot(g100, gaim)
```

Now let's compute a fit between the FT100 and AIM:

```python
np.polyfit(g100, gaim, 1)
```

This implies the following relationship:
    
$g_{aim} =  0.06564199 + 0.93003529 \times g_{100}$
