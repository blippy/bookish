## Introduction

Given X~Beta(a, b), the following is true ([How do I choose parameters for my beta prior?](http://stats.stackexchange.com/questions/64852/how-do-i-choose-parameters-for-my-beta-prior)):

E[X] = a / (a + b),


var[X] = a b / ((a + b)^2 (a + b +1)).


## Plots of the beta distribution

This section displays a series of plots for the beta distribution for various choices of a (alpha) and b (beta). This is useful for the selection of prior distributions.

The plots are generated using the following common code:

```python
%matplotlib inline
import matplotlib.pyplot as plt
import pylab
import numpy
import pandas as pd
from functools import partial
from scipy.stats import beta
```

```python
def plot_beta(a, b, filename = None):
    betap = partial(beta.pdf, a, b)
    xs = numpy.linspace(0,1)
    ys = [beta.pdf(x, a, b) for x in xs]
    plt.plot(xs, ys)
    plt.title("Beta distribution with a = " + str(a) + ", b = " + str(b))
    if filename != None: plt.savefig(filename)
```

```python
plot_beta(a = 0.5, b = 0.5) #, filename = "beta0505.eps")
```

```python
plot_beta(a = 1, b = 1)
```

```python
plot_beta(a = 2, b = 2)
```

```python
plot_beta(a = 10, b = 10)
```

## Probability Mass Function

This section shows how to create a PMF (Probability Mass Function) in both Python and Haskell, using the Beta(5, 3) distrbution over 11 buckets as an example. Firstly, in Python:

```python
xs = numpy.linspace(0,1, 11)
ys_raw = [beta.pdf(x, 5, 3) for x in xs]
ys = [y / sum(ys_raw) for y in ys_raw]
for y in ys: print(y)        
```
