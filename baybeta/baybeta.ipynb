# Bayesian Beta

```python
%matplotlib inline
import matplotlib.pyplot as plt
import pylab
import numpy
import pandas as pd
```

```python
from functools import partial
from scipy.stats import beta
```

```python
def plot_beta(a, b, filename = None):
    betap = partial(beta.pdf, a, b)
    xs = numpy.linspace(0,1)
    ys = [beta.pdf(x, a, b) for x in xs]
    plt.plot(xs, ys)
    plt.title("Beta distribution with a = " + str(a) + ", b = " + str(b))
    if filename != None: plt.savefig(filename)
```

```python
plot_beta(a = 0.5, b = 0.5, filename = "beta0505.eps")
```

```python
plot_beta(a = 1, b = 1, filename = "beta11.eps")
```

```python
plot_beta(a = 2, b = 2, filename = "beta22.eps")
```

```python
plot_beta(a = 10, b = 10, filename = "beta1010.eps")
```
