```python
import h5py
import mython.times
```

```python
with h5py.File("hyh.h5") as h5:
    dstamps  = mython.times.str_to_date(h5["/dstamp"], nparray = True)
    closeprice = np.array(h5["/closeprice"])
    vol = np.array(h5["/vol"])    
```

```python
fig, ax = plt.subplots(nrows=2, ncols=1)
fig.set_size_inches(10, 9)
ax[0].plot(dstamps, closeprice)
ax[1].bar(dstamps, vol)
#plt.show()
plt.savefig('hyh-01.png')
```
