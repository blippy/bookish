# Sorting stochastically

**Abstract** I thought it would be fun to see if "monte carlo" methods could be applied to sorting algorithms. I outline two algorithms that I think are quite promising at improving the sorting time.

---

Needless to say, if you sort elements in an array randomly, then you cannot guarantee order. However, if you can obtain "approximate" order, then it should speed up the running time for strict-ordering algorithms such as quicksort.

To begin my simulation, I need to import a couple of libraries into my notebook:

```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
```

Now let me create an array of random numbers:

```python
nums = np.random.rand(100)
```

I chose 100 numbers, because I want to be able to create sensible plots:

```python
plt.plot(nums)
```

As you can see, the numbers are taken from a uniform distribution in the range [0,1).

Here's what the plot should look like when the array is sorted:

```python
plt.plot(sorted(nums))
```

Let me define a convenience function, ``rsortn``, which, when given a sorting algorithm ``alg`` and an array ``arr``, will apply that algorithm ``n`` times:

```python
def rsortn(alg, arr, n):
    carr = np.copy(arr)
    for i in range(n): carr = alg(carr)
    return carr
```

## Algorithm 1: randomised swapping

The first algorithm I want to choose is the simplest: for as many items as there are in the array, pick two elements at random. If they are in the wrong order, then swap them. Here's the Python code for it:

```python
def alg1(nums):
    nums1 = np.copy(nums)
    for i in range(100):
        i1 = np.random.randint(0,100)
        i2 = np.random.randint(0,100)
        i1, i2 = min(i1,i2), max(i1, i2)
        if(nums1[i1] > nums1[i2]): nums1[i1], nums1[i2] = nums1[i2], nums1[i1]
    return nums1    
```

After one iteration, the plot might look like this:

```python
plt.plot(alg1(nums))
```

You will note that there's some improvement. If the algorithm is repeated 10 times, then the sorting is more convincing:

```python
plt.plot(rsortn(alg1, nums, 10))
```

You can see that there has clearly been some good ordering, with complexity $\mathcal{O}(n)$, where $n$ is the number of points, and I assuming 10 as a constant.

## Algorithm 2: positional placement

In this algorithm, we attempt to use an a priori knowledge/assumption about the distribution of the keys. We know that the array is taken from a uniform distribution, so that given a value $v$ in an array, we can readily compute where it "should" fit in the eventual order. If we knew that they key values had some other distribution, then we could apply an appropriate transformation from key value $v$ to an index value $i$. Maybe the key values are normally distributed, for example, in which case you could apply an inverse Z-transformation. I think it would be rare to say that you knew nothing at all about how the keys were distributed, and would therefore be incapable of proposing any kind of transform between value in index.

So our second algorithm is this:
* loop through each index in the array, which we will call the source index
* apply a transform to obtain a target index and value for the source index
* if the source value and target value is in the wrong order, then swap them. Otherwise, leave them alone

You will notice that the line::

has some slight randomising to it. This is so that the sorting doesn't become "stuck" if we apply the algorithm repeatedly. After one round of iteration, the results are as follows:

## Other ideas

I have had a number of other ideas on using randomisation to help the sorting process:
* Adapt algorithm 2 so that instead of looping through all the indices, you start at index 0, compute a target index, swap values, and then use the old target for the new source. 
* After an iteration, mark the positions where there are "outliers", and then repeat the sorting on these outlying values.
* Alternatively (and probably better): after the first iteration, perform some swapping with items in the first half of the array that are above average with items in the second half of the array that are below average.

All have their complexity of implementations, and are likely to prove a furtile ground for bugs, though.
